```python
from sklearn.preprocessing import LabelEncoder

Datawithoutbill_df = pd.read_csv("Data_withoutBillNo.csv")
# Datawithoutbill_df = Datawithoutbill_df[~Datawithoutbill_df['DEST_CODE'].isna()]
Datawithoutbill_df = Datawithoutbill_df.dropna()

terms_to_remove = ["สินค้าตกค้าง", "เบ็ตเตล็ด",'/']

def clean_item(desc):
    for term in terms_to_remove:
        desc = desc.replace(term, "")
    return desc.strip()  # remove any trailing spaces

def categorize_item(desc):
    if "เอกสาร" in desc or "จดหมาย" in desc:
        return "Document"
    elif "เหมา" in desc:
        return "เหมา"
    elif "สินค้าตกค้าง" in desc:
        return "NotFound"
    elif "เบ็ตเตล็ด" in desc:
        return "Miscellaneous"
    else:
        return "Other"
    
    
label_encoder = LabelEncoder()

Datawithoutbill_df['Category'] = Datawithoutbill_df['ITEM_DESCแสดงหน้าบิล'].apply(categorize_item)
Datawithoutbill_df['ITEM_DESC_CLEAN'] = Datawithoutbill_df['ITEM_DESCแสดงหน้าบิล'].apply(clean_item)
# Datawithoutbill_df['ITEM_DESC_SPLIT'] = Datawithoutbill_df['ITEM_DESC_CLEAN'].str.split('/|,|\+')

Datawithoutbill_df['BILL_DATE'] = pd.to_datetime(Datawithoutbill_df['BILL_DATE'], format='%Y%m%d')
# Datawithoutbill_df['year'] = Datawithoutbill_df['BILL_DATE'].dt.year
Datawithoutbill_df['month'] = Datawithoutbill_df['BILL_DATE'].dt.month
Datawithoutbill_df['day'] = Datawithoutbill_df['BILL_DATE'].dt.day
Datawithoutbill_df['day_of_week'] = Datawithoutbill_df['BILL_DATE'].dt.dayofweek
Datawithoutbill_df['day_of_year'] = Datawithoutbill_df['BILL_DATE'].dt.dayofyear
Datawithoutbill_df['quarter'] = Datawithoutbill_df['BILL_DATE'].dt.quarter
Datawithoutbill_df['is_weekend'] = Datawithoutbill_df['BILL_DATE'].dt.dayofweek >= 5

Datawithoutbill_df['ITEM_DESC_CLEAN_labelencoding'] = label_encoder.fit_transform(Datawithoutbill_df['ITEM_DESC_CLEAN'])
Datawithoutbill_df['CONSIGNEE_TAMBON'] = label_encoder.fit_transform(Datawithoutbill_df['CONSIGNEE_TAMBON'])
Datawithoutbill_df['CONSIGNEE_AMPHUR'] = label_encoder.fit_transform(Datawithoutbill_df['CONSIGNEE_AMPHUR'])
# Datawithoutbill_df['CONSIGNEE_PROVINCE'] = label_encoder.fit_transform(Datawithoutbill_df['CONSIGNEE_PROVINCE'])
Datawithoutbill_df['ITEM_NAMEในระบบ'] = label_encoder.fit_transform(Datawithoutbill_df['ITEM_NAMEในระบบ'])
Datawithoutbill_df['UNIT_NAME'] = label_encoder.fit_transform(Datawithoutbill_df['UNIT_NAME'])
Datawithoutbill_df['Category'] = label_encoder.fit_transform(Datawithoutbill_df['Category'])
Datawithoutbill_df['REG_Code'] = label_encoder.fit_transform(Datawithoutbill_df['REG_Code'])
Datawithoutbill_df['DEST_CODE'] = label_encoder.fit_transform(Datawithoutbill_df['DEST_CODE'])

# Datawithoutbill_df['QTY'] = int(Datawithoutbill_df['QTY'])
Datawithoutbill_df['QTY'] = pd.to_numeric(Datawithoutbill_df['QTY'], errors='coerce')

# Datawithoutbill_df = Datawithoutbill_df.drop(['BILL_DATE','BILL_INDEX','ITEM_DESC_CLEAN','ITEM_DESCแสดงหน้าบิล'],axis=1)
# 
Datawithoutbill_df
```

```python
# ============================================================



Running_arry = list_array[:1000]


print(Running_arry)
print(len(Running_arry))



# ============================================================
```

```python
weather_cache = {}

def convert_to_numeric(column):
    return column.replace({r'°F': '', r'°%': '', r'°mph': '', r'°in': '', r'mi': ''}, regex=True).astype(float)


def scrape_weather_data(date_: str,Province , driver_path: str = None) -> pd.DataFrame:
    if date_ in weather_cache:
        return weather_cache[date_]
    
    URL = Weather_map[Weather_map['Province'] == Province]['URL'].values[0]


    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    driver.get(f'{URL}/date/{date_}')

    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'table.mat-table'))
        )
    except:
        driver.quit()
        return None

    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')

    table = soup.find('table', {'class': 'mat-table cdk-table mat-sort ng-star-inserted', 'role': 'grid'})

    if table:
        rows = table.find_all('tr')

        headers = [header.text.strip() for header in rows[0].find_all('th')]

        data = []
        for row in rows[1:]:
            cells = [cell.text.strip() for cell in row.find_all('td')]
            data.append(cells)

        # Create a DataFrame
        df = pd.DataFrame(data, columns=headers)
        
        df['Temperature'] = convert_to_numeric(df['Temperature'])
        df['Dew Point'] = convert_to_numeric(df['Dew Point'])
        df['Humidity'] = convert_to_numeric(df['Humidity'])
        df['Wind Speed'] = convert_to_numeric(df['Wind Speed'])
        df['Wind Gust'] = convert_to_numeric(df['Wind Gust'])
        df['Pressure'] = convert_to_numeric(df['Pressure'])
        df['Precip.'] = convert_to_numeric(df['Precip.'])
        
        numeric_means = df[['Temperature', 'Dew Point', 'Humidity', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']].mean()
        categorical_modes = df[[ 'Wind', 'Condition']].mode().iloc[0]
        numeric_means['Wind'] = categorical_modes['Wind']
        numeric_means['Condition'] = categorical_modes['Condition']
        
        # Cache the scraped data for future use
        weather_cache[date_] = numeric_means
        driver.quit()
        return numeric_means

    else:
        # print("Table not found.")
        driver.quit()
        return None
```

```python

Time = {
    'BILL_DATE': [],
    'CONSIGNEE_PROVINCE': [],
    'Temperature': [],
    'Humidity': [],
    'Wind_Speed': [],
    'Condition': [],
    'Dew_Point' : [],
    'Wind': [],
    'Wind_Gust': [],
    'Pressure': [],
    'Precip': [],
}

for i in tqdm(range(len(Running_arry))):
# for i in tqdm(range(5)):

    DATE_time , Province = Running_arry[i]

    Feature = scrape_weather_data(date_=DATE_time,Province=Province)
    
    if Feature is not None:
        # Append the date
        Time['BILL_DATE'].append(DATE_time)
        Time['CONSIGNEE_PROVINCE'].append(Province)
        Time['Temperature'].append(Feature.Temperature)  # Clean °F
        Time['Humidity'].append(Feature.Humidity)  # Clean %
        Time['Wind_Speed'].append(Feature['Wind Speed'])  # Clean mph
        Time['Condition'].append(Feature.Condition)  # No need to clean text data
        Time['Dew_Point'].append(Feature['Dew Point'])  # Clean °F
        Time['Wind'].append(Feature.Wind)  # Clean %
        Time['Wind_Gust'].append(Feature['Wind Gust'])  # Clean mph
        Time['Pressure'].append(Feature.Pressure)  # No need to clean text data
        Time['Precip'].append(Feature['Precip.'])  # No need to clean text data
    else:
        # If the weather data wasn't available, append None or a default value
        Time['BILL_DATE'].append(DATE_time)
        Time['CONSIGNEE_PROVINCE'].append(Province)
        Time['Temperature'].append('N/A')
        Time['Humidity'].append('N/A')
        Time['Wind_Speed'].append('N/A')
        Time['Condition'].append('N/A')
        Time['Dew_Point'].append('N/A')  # Clean °F
        Time['Wind'].append('N/A')  # Clean %
        Time['Wind_Gust'].append('N/A')  # Clean mph
        Time['Pressure'].append('N/A')  # No need to clean text data
        Time['Precip'].append('N/A')  # No need to clean text data

# Convert the dictionary to a pandas DataFrame for better data handling
df = pd.DataFrame(Time)

# df['Temperature'] = df['Temperature'].apply(lambda x: float(x.replace('\xa0°F', '')))
# df['Humidity'] = df['Humidity'].apply(lambda x: float(x.replace('\xa0°%', '')))
# df['Wind_Speed'] = df['Wind_Speed'].apply(lambda x: float(x.replace('\xa0°mph', '')))
# df['Dew_Point'] = df['Dew_Point'].apply(lambda x: float(x.replace('\xa0°F', '')))
# df['Wind_Gust'] = df['Wind_Gust'].apply(lambda x: float(x.replace('\xa0°mph', '')))
# df['Pressure'] = df['Pressure'].apply(lambda x: float(x.replace('\xa0°in', '')))
# df['Precip'] = df['Precip'].apply(lambda x: float(x.replace('\xa0°in', '')))

print(df)
```
